```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,5,7,9,17,2,3,5])
plt.show()
```

```python
plt.plot([-3,-1,0,4,7],[1,4,6,8,17])
plt.show()
```

```python
plt.plot([-3,-1,0,4,7],[1,4,6,8,17])
plt.axis([-4,8,0,10])
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("square function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x **2
y2 = x+1 
plt.plot(x,y,'b--',x,y2,'g')
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x **2
y2 = x+1 
plt.plot(x,y,'b--',label="x**2")
plt.plot(x,y2, 'g',label="x+1")
plt.legend(loc="best")
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x **2
y2 = x+1 
plt.subplot(1,2,1)
plt.plot(x,y, 'b--')
plt.subplot(1,2,2)
plt.plot(x,y2, 'g')
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x **2
y2 = x+1 
plt.subplot(1,2,1)
plt.plot(x,y, 'b--')
plt.xlabel("x1",fontsize=10)
plt.ylabel("y1", fontsize=10)
plt.subplot(1,2,2)
plt.plot(x,y2, 'g')
plt.xlabel("x2",fontsize=10)
plt.ylabel("y2", fontsize=10)
plt.show()
```

```python
from numpy.random import rand
x,y = rand(2,100)
plt.scatter(x,y)
plt.show()
```

```python
from numpy.random import rand
x,y = rand(2,100)
x2,y2 = rand(2,100)
plt.scatter(x,y, c='red')
plt.scatter(x2,y2, c='blue')
plt.show()
```

```python
data = [1,1.1,1.8,2,2.1,3.2,3,3,3,3]
plt.subplot(211)
plt.hist(data, bins=10,rwidth=0.8)
plt.subplot(212)
plt.hist(data,bins=[1,1.5,2,2.5,3], rwidth=0.95)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
